# Long-short term memory
Long-Short Term Memory networks or LSTMs are a variant of RNN that solve the Long term memory problem. LSTM is also an artificial recurrent neural network architecture used in the field of deep learning. Unlike standard feedforward neural networks, LSTM has feedback connections. It can not only process single data points, but also entire sequences of data.

# Fake News Classification using Long-short term memory
[Dataset](https://www.kaggle.com/c/fake-news/data#)

```python
#importing necessary library
import pandas as pd
import tensorflow as tf
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
from tensorflow.keras.preprocessing import text
```

```python
#loading the dataset
df=pd.read_csv('/content/train.csv')
df.head()
```

```python
#Drop Nan Values
df=df.dropna()
```

```python
#Get the Independent Features
X=df.drop('label',axis=1)
```

```python
#Get the Dependent features
y=df['label']
```

```python
#checking out the x shape
X.shape
```

```python
#also checking the y shape
y.shape
```

```python
#Vocabulary size
voc_size=5000
```

```python
#defifning one shot representation
messages=X.copy()
messages['title'][1]
```

```python
messages.reset_index(inplace=True)
```

```python
#importing required library
import nltk
import re
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
#Preprocessing dataset
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
corpus = []
for i in range(0, len(messages)):
    print(i)
    review = re.sub('[^a-zA-Z]', ' ', messages['title'][i])
    review = review.lower()
    review = review.split()
    
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
#definfing oneshot representation
onehot_repr = [text.one_hot(word,voc_size) for word in corpus]
onehot_repr
```

```python

#Embedding Representation
sent_length=20
embedded_docs=pad_sequences(onehot_repr,padding='pre',maxlen=sent_length)
print(embedded_docs)
```

```python
embedded_docs[0]
```

```python
# Defining model
embedding_vector_features=40
model=Sequential()
model.add(Embedding(voc_size,embedding_vector_features,input_length=sent_length))
model.add(LSTM(100))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
print(model.summary())
```

```python
len(embedded_docs),y.shape
```

```python
#storing the final output
import numpy as np
X_final=np.array(embedded_docs)
y_final=np.array(y)
```

```python
#printing fianl x & y shape
X_final.shape,y_final.shape
```

```python
#spliting the dataset into test and train
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_final, y_final, test_size=0.33, random_state=42)
```

```python
#Model training
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=10,batch_size=64)
```

```python
#adding dropout
from tensorflow.keras.layers import Dropout
#Creating model
embedding_vector_features=40
model=Sequential()
model.add(Embedding(voc_size,embedding_vector_features,input_length=sent_length))
model.add(Dropout(0.3))
model.add(LSTM(100))
model.add(Dropout(0.3))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
#performance matrix
y_pred=model.predict(X_test)
from sklearn.metrics import confusion_matrix
```

```python
#printing confusion matrix
confusion_matrix(y_test,y_pred.round())
```

```python
#printing the accuracyS
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred.round())
```
