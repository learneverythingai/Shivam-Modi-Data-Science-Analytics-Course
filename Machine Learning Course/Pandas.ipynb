### Import Pandas

### What is a Series?
A Pandas Series is like a column in a table.

It is a one-dimensional array holding data of any type.

```python
# Create a simple Pandas Series from a list:
a = [1, 2, 3, 4]
series = pd.Series(a)
print(series)
```

### Create Labels
With the index argument, you can name your own labels.

```python
series = pd.Series(a, index = ["x", "y", "z","a"])
print(series)
```

When you have created labels, you can access an item by referring to the label.

```python
# check the value at index "a"
series["a"]
```

Create a simple Pandas Series from a dictionary of temperature: {"day1": 42, "day2": 40, "day3": 39}

```python
temperature = {"day1": 42, "day2": 40, "day3": 39}
series = pd.Series(temperature)
print(series)
```

## DataFrame Operations

### Create new columns for your dataframe

```python
# create a series: cities ["Chandigarh","Delhi","Kanpur","Chennai","Manali","Mumbai","Ahmedabad"] 
city = pd.Series(["Chandigarh","Delhi","Kanpur","Chennai","Manali","Mumbai","Ahmedabad"])
city
```

```python
# add this column series in your old dataframe df and name it as "cities"
df["cities"] = city
df.head()
```

### mean, max, min, median
you can just output any of the above operations by calling method. i.e, dataframe['column_name'].mean() to calculate the mean of the column 

```python
# print mean, max, min and median
print(df['temperature'].mean())
print(df['temperature'].max())
print(df['temperature'].min())
print(df['temperature'].median())
```

Suppose you want to check which rows have temperature greater than mean temperature of column or which rows have temperature less than 40. In these cases, we can apply conditions on pandas dataframe. 

dataframe[condition]. Example, to check on which days we have temperature greater than 40, we will write: df[df['temperature']>40]["weekday"]. Here, df[df['temperature']>40] is a dataframe that satisfies the condition.

```python
# output the above example
df[df['temperature']>40]["weekday"]
```

```python
# check the temperature on monday
df[df["weekday"]=='monday']['temperature']
```

```python
# check the weekdays when temperature was more than mean temperature
df[df['temperature']>df['temperature'].mean()]['weekday']
```

### Drop a column
We found that weekday column is irrelevant to us so drop this column, so what if we want to drop any column? 

For that, we will write dataframe.drop('column name',axis=1,inplace=True). Here, we need to define axis=1, which is called axis for columns because we must define its axis(=1) to drop any column. inplace=True is used to specify to commit that changes in the dataframe as well or else you'll have to assign a new dataframe to have such changes. 

```python
# drop the weekday column
df = df.drop('weekday',axis=1) # or df.drop('weekday',axis=1,inplace=True)
df.head()
```

Now suppose we want to add some more rows in our dataframe. So, how do we merge two dataframes? 

For this task, we will use pandas concat function.

d3 = pd.cocat[d1,d2]

Example, Create another datafeame 'd2' where the columns are
'cities' - ['Bengalaru','kolkata','Srinagar','Pune']
'temperature' - [34,35,36,39] 'wind' - ['slow','slow','fast','fast'] 'weather' - ['cold','cold','sunny','cloudy']

## Pandas Index
Pandas is a best friend to a Data Scientist, and index is the invisible soul behind pandas. 

Index is like an address, thats how any data point across the dataframe or series can be accessed. Rows and columns both have indexes, rows indices are called as index and for columns its general column names.

Let's check out index of our dataframe then.

```python
# print the index of dataframe df
print(df.index)
```

Well we have learnt that index is like soul for dataframe and it's like an address for each row but here we see that indexes are repeating. Index should have unique values.

Well it is because we merged two dataframes so together their indexes were also merged. Indexing in pandas means simply selecting particular rows and columns of data from a DataFrame. And, Indexes are very important for our algorithms thus we must never miss out to look at them as they are very important part for data preprocessing as well. So, we keep unique values in the index. 

Now, suppose we want to look everything in perspective of cities, so I would want to make cities as my index column. We can use dataframe.set_index('column_name',inplace=True) method to do this. 

```python
# set cities column as index of df and print the df index 
df.set_index('cities',inplace=True)
df.index
```

```python
# print df
df
```

Now, you can see that we have set cities column as our index. Generally, such operations are done in date or time columns of certain datasets because it is more convinient to use date/time as index in Machine Learning models.

You can also reset the dataframe index using reset_index function. reset_index in pandas is used to reset index of the dataframe object to default indexing (0 to number of rows minus 1) or to reset multi level index. By doing so, the original index gets converted to a column. 

```python
# reset df index
df.reset_index(inplace=True)
df.head()
```

```python
# print the index of df
print(df.index)
```

Now, you can see that index of df is shown as RangeIndex(start=0, stop=11, step=1). It means that our index starts from 0, with step of 1, which is increasing by 1 till it gets to 11. 

## Apply functions in Pandas
Use apply() function when you wanted to update every row in pandas DataFrame by calling a custom function.

Example: Create a function that defines temperature in categories: low, moderate and high. If temperature is less than 36, then low, if between 36 and 40, then moderate and more than 40 then high.

```python
# create the function
def convert_to_classes(x):
  if x < 36:
    return 'low'
  if x >= 36 and x <= 40:
    return 'moderate' 
  else:
    return 'high'
```

```python
df['temperature_category'] = df['temperature'].apply(convert_to_classes)
```
